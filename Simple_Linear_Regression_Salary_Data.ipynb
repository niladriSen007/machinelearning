```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Salary_dataset.csv')
x = data.iloc[:,0:1].values
y = data.iloc[:,1:2].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
years_of_exp_train_data_predict_val =  regressor.predict(x_train)
years_of_exp_test_data_predict_val =  regressor.predict(x_test)
```

```python
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,years_of_exp_train_data_predict_val,color="blue")
plt.title("Salary vs Experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x_test,y_test,color="violet")
plt.plot(x_train,years_of_exp_train_data_predict_val,color="blue")
plt.title("Salary vs Experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```
